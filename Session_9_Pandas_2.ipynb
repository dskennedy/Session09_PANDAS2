```python
sess9_df=pd.read_csv('C:\\Users\\User\\US_Baby_Names_right.csv')
```

```python
#Removing un-named column
sess9_df=sess9_df.drop(sess9_df.columns[sess9_df.columns.str.contains('Unnamed', case=False)], axis=1)
sess9_df
```

```python
# Gender distribution on yearly basis
sess9_df.groupby(['Year','Gender'])['Gender'].count().unstack()
```

```python
sess9_df.groupby('Gender').Year.plot(kind='kde')
```

```python
# Top 5 names.
sess9_df.groupby('Name')['Name'].count().sort_values(ascending=False).head(5)
```

```python
# State wise distribtution
sess9_df.groupby(['State','Gender'])['Gender'].count().unstack()
```
